# Question 3:
In this question we are going to classify data generated using a normal distribution with an adaline neuron.
Our dataset consists of 3 columns in which the first 2 are features and the last one shows class of each datapoint.

First of all, we import required libraries:

```python
import matplotlib.pyplot as plt
import numpy as np
```

Now, we generate th first data using normal distribution and specified variance and mean:

```python
np.random.seed(13)
x1 = np.random.normal(1, 0.5, 1000)
y1 = np.random.normal(1, 0.5, 1000)
x2 = np.random.normal(-1, 0.5, 10)
y2 = np.random.normal(-1, 0.5, 10)
```

```python
temp = np.ones(1000)
temp2 = np.ones(10)
temp2 = -1 * temp2
data = np.concatenate((np.c_[x1, y1, temp], np.c_[x2, y2, temp2]))
```

We can see that our generated data looks exactly as expected.

```python
for i in range(len(data)):
    datapoint = data[i]
    if datapoint[2] == 1:
        color = "g"
    if datapoint[2] == -1:
        color = "r"
    plt.scatter(datapoint[0], datapoint[1], c = color, s = 7)
plt.title('Generated data and their classes')
plt.xlabel('x1')
plt.ylabel('y1')
```

In the next step, we set the values of weight and biases randomly. Also learning rate is specified here.

```python
learning_rate = 0.1
np.random.seed(13)
w1 = np.random.randn()
w2 = np.random.randn()
b = np.random.randn()
```

In this step, we learn dataset using the algorithms and updating rules taught in the class. I used 10 epoches but fewer epochs could have done the job.

```python
for epoch in range(10):
    for i in range(1000 + 10):
        x = data[i]
        e = x[2] - np.tanh(x[0] * w1 + x[1] * w2 + b)
        w1 = w1 + learning_rate * e * x[0]
        w2 = w2 + learning_rate * e * x[1]
        b = b + learning_rate * e
```

Now we draw the line between two classes and it can be seen from the plot below tht we have achieved 100 percent accuracy.

```python
for i in range(len(data)):
    datapoint = data[i]
    if datapoint[2] == 1:
        color = "g"
    if datapoint[2] == -1:
        color = "r"
    plt.scatter(datapoint[0], datapoint[1], c = color, s = 7)

x = np.linspace(-4, 4)
y = ((-w1 * x) - b) / w2
plt.plot(x, y , c = 'black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Classified data using adaline neuron')
```

Now, we generte second dataset like last example using normal distributions and specified varianced and means.

```python
x1 = np.random.normal(1, 0.5, 100)
y1 = np.random.normal(1, 0.5, 100)
x2 = np.random.normal(-1, 0.5, 100)
y2 = np.random.normal(-1, 0.5, 100)
temp = np.ones(100)
temp2 = np.ones(100)
temp2 = -1 * temp2
data2 = np.concatenate((np.c_[x1, y1, temp], np.c_[x2, y2, temp2]))
for i in range(len(data2)):
    datapoint = data2[i]
    if datapoint[2] == 1:
        color = "g"
    if datapoint[2] == -1:
        color = "r"   
    plt.scatter(datapoint[0], datapoint[1], c = color, s = 7)
plt.title('Generated data and their classes')
plt.xlabel('x1')
plt.ylabel('y1')
```

And like the last example we specify weights an biases and learning rate.

```python
learning_rate = 0.1
w1 = np.random.randn()
w2 = np.random.randn()
b = np.random.randn()
```

```python
for epoch in range(10):
    for i in range(200):
        x = data2[i]
        net = x[0] * w1 + x[1] * w2 + b
        target = x[2]
        e = target - np.tanh(net)
        w1 = w1 + learning_rate * e * x[0]
        w2 = w2 + learning_rate * e * x[1]
        b = b + learning_rate * e
```

It can be seen that our neuro was able to classify in this example too.
