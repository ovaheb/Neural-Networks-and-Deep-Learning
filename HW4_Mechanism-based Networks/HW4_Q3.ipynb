## Importing required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import copy
```

## Algorithm Implementation

```python
def activation(x):
    if x < 0:
        return 0
    elif x >= 0 and x < 2:
        return x
    else:
        return 2

def mexican_hat(x, R1, R2, C1, C2, t_max):
    data = []
    x_i = copy.deepcopy(x)
    x_old_i = x_i
    for t in range(t_max):
        for i in range(len(x)):
            x_i[i] = C1 * np.sum(x_old_i[max(0, i - R1) : min(len(x) - 1, i + R1 + 1)]) 
            x_i[i] += C2 * np.sum(x_old_i[max(0, i - R2) : max(0, i - R1)])
            x_i[i] += C2 * np.sum(x_old_i[min(len(x) - 1, i + R1 + 1) : min(len(x) - 1, i + R2 + 1)])
        x_old_i = []
        for i in range(len(x)):
          x_old_i.append(activation(copy.deepcopy(x_i)[i]))
        x_old_i = np.array(x_old_i)
        data.append(x_old_i)
    return data
```

```python
def plot_res(data):
    plt.figure(figsize = (10, 10))
    for i in range(len(data)):
        plt.plot(data[i], label = 't = ' + str(15 - i))
    plt.legend()
    plt.title('Mexican hat result during trining')
    plt.xlabel('x_i')
    plt.ylabel('x')
    plt.show() 
    plt.figure(figsize = (10, 10))
    plt.plot(data[len(data) - 1])
    plt.title('Mexican hat final result')
    plt.xlabel('x_i')
    plt.ylabel('x')
    plt.show()
```

## Result with R1 = 0 and R2 = 1000

```python
x = [0.32, 0.33, 0.28, 0.47, 0.66, 0.80, 0.4, 0.33, 0.1 , 0.26]
x = np.array(x)
R1 = 0
R2 = 1000
C1 = 0.9
C2 = -0.2
t_max = 15
data1 = mexican_hat(x, R1, R2, C1, C2, t_max)
plot_res(data1)
```

## Result with R1 = 1 and R2 = 5

```python
x = [0.32, 0.33, 0.28, 0.47, 0.66, 0.80, 0.4, 0.33, 0.1 , 0.26]
x = np.array(x)
R1 = 1
R2 = 5
C1 = 0.3
C2 = -0.3
t_max = 15
data2 = mexican_hat(x, R1, R2, C1, C2, t_max)
plot_res(data2)
```
